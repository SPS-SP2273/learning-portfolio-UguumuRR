```python
import os
import shutil

city_names = ['Dhaka', 'Jakarta', 'Karachi', 'Manila',
              'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

lower_list = [word.lower() for word in city_names]

city_files_dict = {}

# Populate city_files_dict with files for each city
for city in lower_list:
    city_files_dict[f'{city}_files'] = glob.glob(f'os-collaboration-exercise-data/*{city}*')

# Copy files for each city to its corresponding folder
for city, files in city_files_dict.items():
    destination_folder = os.path.join('Cities', city[:-6])  # Extract city name from the variable name
    os.makedirs(destination_folder, exist_ok=True)  # Ensure destination folder exists
    for file in files:
        shutil.copy(file, destination_folder)
        #print(f'Copied {file} to {destination_folder}')
```
