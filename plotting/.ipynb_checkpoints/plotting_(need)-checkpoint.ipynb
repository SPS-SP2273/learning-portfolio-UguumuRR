# A simple plot

```python
from matplotlib import pyplot as plt
```

## Letâ€™s look at some code.

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_1, label='Y values',
         color='red', linestyle='dashed')
plt.xlabel('x-values') #names the horizontal axis 
plt.ylabel('y-values') #names the vertical axis 
plt.title('X vs Y')
plt.grid(alpha=.5) #the opacity of the grid  
plt.legend(loc='upper left') #where the label will be placed
```

### Things to note

1.
| Long form | Abbreviation |
| :--- | :--- |
| color | c |
| linestyle | ls |
| linewidth | lw |
2. Although the plot can be shown in Jupyter without __plt.show()__, it is a good practice to include since it enables direct calling from the command line.
3. The plotting functions usually have the default values for the styling parameters. 


```python
plt.plot(x, y_1, y_2)
```

4. The order of keyword arguments (color, linewidth) doesn't matter. 

## Adding another plot

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Lets start plotting
plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed')
plt.plot(x, y_2, label='Y2 values', color='blue')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.5)
plt.legend(loc='upper left')
```

## Yet another plot but with error bars

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75 #this is how you will manipulate the values. 

# Lets start plotting
plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed',)
plt.plot(x, y_2, label='Y2 values', color='blue', )
plt.errorbar(x, y_3, xerr=x_error, yerr=y_error,
             label='Y3 with errors', color='green')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

# Better with NumPy (cause arrays have element-wise operations)

```python
import numpy as np
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

np_x = np.array(x)
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

plt.plot(np_x, np_y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(np_x, np_y_2, color='blue', label='Y2 values')
```

## Adding mathematical functions

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])

x2 = x**2                  # The math stuff
cos_x = np.cos(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed', )
plt.plot(x, cos_x, label='cos(x)',
         color='blue')
plt.legend()                                 
```

## We need another axis!

Matplotlib offers a variety of ways to have multiple axes. The simplest way is to have another y-axis that shares the same x-axis. We can use the command __twinx()__ for this.

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
cos_x = np.cos(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  # For y-axis 1

plt.twinx()                                                   # This creates a new y-axis 
                                                              # for the plots that comes after
plt.plot(x, cos_x, label='cos(x)',color='blue', )
plt.legend(loc='lower right')                                 # For y-axis 2
```

```python
#To make the graph more presentable, use np.linspace
x = np.linspace(0, 6, 100)
x2 = x**2
cos_x = np.cos(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  # For y-axis 1

plt.twinx()                                                   # This creates a new y-axis 
                                                              # for the plots that comes after
plt.plot(x, cos_x, label='cos(x)',color='blue', )
plt.legend(loc='lower right')                                 # For y-axis 2
```

# Saving to disc

To export plots at different resolutions in various formats: 
- Use __savefig()__ function. 
- Specify the format with the extension (filename.pdf)
- Specify resolution by use dots-per-inch (dpi) (inch is weird, SI is superior)

```python
x = np.linspace(0, 6, 100)
x2 = x**2
cos_x = np.cos(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  # For y-axis 1

plt.twinx()                                                   # This creates a new y-axis 
                                                              # for the plots that comes after
plt.plot(x, cos_x, label='cos(x)',color='blue', )
plt.legend(loc='lower right') 
plt.savefig('simple-plot.pdf', dpi = 250)
```

# A real example: Global Warming

## Plotting data from files

```python
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.ylim([-8, 8])
```

```python
#To add a horizontal line at the zero value:
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.ylim([-8, 8])
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
```

## Styles

Matplotlib offers some standard style templates such as fivethirtyeight. 

```python
plt.style.use('fivethirtyeight')

data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)

date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
plt.ylim([-8, 8])

plt.xlabel('Date')
plt.ylabel('Temperature Anomaly')
plt.title('Temperature anomaly\n(Relative to  average from Jan 1951 - Dec 1980.)')
```
